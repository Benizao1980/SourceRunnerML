# Data Acquisition

Notes
* To download data generated by PubMLST, use `wget` followed by the URL of the data file.
* To compressed TSV files, use `gzip`. This will reduce the file size and make it easier to transfer.
* Download data jobs on PubMLST are limit to 25,000,000 datapoints, which are calculated as isolates X fields (typing data columns). You need to break down the list of ids to be able to download the data to then merge it into a single `.tsv.gz` file
Routine to break down list of PubMLST ids into a fixed number of subfiles

## C. jejuni

### PubMLST
Search criteria:
* species: Campylobacter jejuni
* source: NOT null
* source: NOT contain "human"

Results:
* Isolate count: 46,965
* Genome count: 28,950
* Date: 18 July 2025

Routine to break down list of PubMLST ids into a fixed number of subfiles

```python
import os 
workdir = '/home/broncio/oxford/source-attribution/SourceRunnerML/data/c_jejuni/misc'
ids_filename = 'CJ_PubMLST_ids_18-Jul-2025.txt'
n_output_files = 4

with open(os.path.join(workdir, ids_filename), 'r') as f:
    ids = f.read().splitlines()

chunk_size = len(ids) // n_output_files
for i in range(n_output_files):
    start = i * chunk_size
    end = (i + 1) * chunk_size if i < n_output_files - 1 else len(ids)
    with open(os.path.join(workdir, f'CJ_PubMLST_ids_18-Jul-2025_part{i+1}.txt'), 'w') as f:
        f.writelines('\n'.join(ids[start:end]))
```
